```python
new_books_def= books[['id', 'title', 'author', 'book_link', 'genre', 'num_of_page', 'rating_count', 'rate']]
```

```python
new_books_def.dropna(inplace=True)
new_books_def.head()
```

```python
#appling steming (converts activity, activities as same word)
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]  
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_books_def['genre']=new_books_def['genre'].apply(stem)
```

```python
new_books_def['genre']= new_books_def['genre'].apply(lambda x: x.lower())
```

```python
#new_books_def['genre']=new_books_def['genre'].apply(lambda x:x.split())
new_books_def.head()
```

```python
for i in new_books_def.index:
    new_books_def['genre'][i] = new_books_def['genre'][i].split("-")
    print(new_books_def['genre'][i])
```

```python
new_books_def['genre']= new_books_def['genre'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
new_books_def = new_books_def[new_books_def['genre'].notnull()]
#new_books_def['genre']=new_books_def['genre'].apply(lambda x:x.split())
```
